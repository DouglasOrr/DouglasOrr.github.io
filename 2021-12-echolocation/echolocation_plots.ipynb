```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context("talk")
```

```python
plt.figure(figsize=(8, 4.5))
xs = np.arange(0, 4e-2 + 1e-4, 1e-5)
ys = np.sin(700 * xs * 2 * np.pi) * np.sin(xs * 2 * np.pi / 0.08)
plt.plot(xs, ys, label=r"$\mathrm{sin}(700 \cdot 2 \pi \, t) \cdot \mathrm{sin}(\frac{1}{0.08} \cdot 2 \pi \, t)$")
plt.xlabel("t /s")
plt.title(r"$\mathrm{sin}(700 \cdot 2 \pi \, t) \cdot \mathrm{sin}(\frac{1}{0.08} \cdot 2 \pi \, t)$",)
plt.yticks([-1, 0, 1])
plt.xticks([0, 0.01, 0.02, 0.03, 0.04])
plt.hlines(0, xs.min(), xs.max(), "k", lw=0.5)
plt.xlim(xs.min(), xs.max())
for direction in ["top", "right"]:
    plt.gca().spines[direction].set_visible(False)
```

```python
plt.figure(figsize=(4, 3))
xs = np.arange(0, 1, 1e-3)
ys = np.sin(4 * xs * 2 * np.pi)
plt.plot(xs, ys, lw=4)
plt.axis("off");
```

```python
plt.figure(figsize=(4, 3))
xs = np.arange(0, 1, 1e-3)
ys = np.sin(xs * np.pi)
c0 = np.array(sns.color_palette()[0])
plt.fill(xs, ys, color=c0, alpha=.5)
plt.fill(xs, -ys, color=np.minimum(1, c0 + .2), alpha=.5)
plt.axis("off");
```
